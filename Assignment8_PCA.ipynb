```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
# import dataset
wine=pd.read_csv('/content/wine.csv')
wine
```

```python
wine['Type'].value_counts()
```

```python
wine.describe()
```

```python
wine.info()
```

```python
#consider only numeric data
wine1=wine.iloc[:,1:]
wine1
```

```python
wine1.info()
```

```python
#converting data in to numpy array
wine_ary=wine1.values
wine_ary
```

```python
std=StandardScaler() # normalize numeric data
wine1_norm=std.fit_transform(wine_ary)
wine1_norm
```

PCA implementation

```python
# Applying PCA Fit Transform to dataset
pca=PCA(n_components=13)

wine_pca=pca.fit_transform(wine1_norm)
wine_pca
```

```python
pca.components_ #PCA component matrix /co variance
```

```python
# The amount of variance that each PCA explains is
pca.explained_variance_

```

```python
# in percentage-The amount of variance that each PCA explains is
var=pca.explained_variance_ratio_
var
```

```python
#cumulative variance
var1=np.cumsum(np.round(var,decimals=4)*100)
var1  
```

```python
# Variance plot for PCA components
plt.plot(var1,color='blue')  
```

```python
#final dataframe

final_df=pd.concat([wine['Type'],pd.DataFrame(wine_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
#Visualization of PCA
import seaborn as sns

fig=plt.figure(figsize=(10,8))
sns.scatterplot(data=final_df)
```

```python
sns.scatterplot(data=final_df, x='PC1', y='PC2', hue='Type');
```

Check with clustering algorithm


Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
# create dendrogram
dendrogram = sch.dendrogram(sch.linkage(wine1_norm, method='complete')) # complete linkage
```

```python
#create clusters=3
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
hc
```

```python
y=pd.DataFrame(hc.fit_predict(wine1_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
#adding clusters to dataset
wine2=wine.copy()
wine2['clustersid']=hc.labels_
wine2
```

Kmean Clustering

```python
from sklearn.cluster import KMeans
```

```python
wcss = [] # within cluster sum of square. Initialize WCSS and begin the loop
for i in range(1,5 ):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=30,random_state=2)
    kmeans.fit(wine1_norm) 
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,5), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

build cluster algorithm  using K=3

```python
kmeans = KMeans(n_clusters=3,random_state=30) # pass no. of clusters as 3
predict=kmeans.fit_predict(wine1_norm)
```

```python
predict
```

```python
#adding clusters to dataset
wine3=wine.copy()
wine3['clustersid']=predict
wine3
```
