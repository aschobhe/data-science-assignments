```python
book1=book.drop('Unnamed: 0',axis=1)
book1
```

```python
#Renaming columns
book1.columns=['Userid','BookTitle','BookRating']
```

```python
book1
```

```python
book1=book1.sort_values(by=['Userid'])
```

```python
#number of unique users in dataset
len(book1.Userid.unique())
```

```python
#number of unique books in dataset
len(book1.BookTitle.unique())
```

```python
book1.BookRating.value_counts()
```

```python
book1.loc[book1['BookRating']=='small','BookRating']=0
book1.loc[book1['BookRating']=='large','BookRating']=1
```

```python
book1_df=book1.pivot_table(index='Userid',columns='BookTitle',values='BookRating').reset_index(drop=True)
```

```python
book1_df.fillna(0,inplace=True)
book1_df
```

Calculating cosine similarity between users

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
user_sim=1-pairwise_distances(book1_df.values,metric="cosine")
```

```python
user_sim
```

```python
user_sim_df=pd.DataFrame(user_sim)
```

```python
user_sim_df
```

```python
#Set the index and column names to user ids 
user_sim_df.index = book1.Userid.unique()
user_sim_df.columns = book1.Userid.unique()
```

```python
user_sim_df
```

```python
np.fill_diagonal(user_sim,0)
user_sim_df
```

```python
#Most Similar Users
print(user_sim_df.idxmax(axis=1)[1348])
print(user_sim_df.max(axis=1).sort_values(ascending=False).head(50))
```
